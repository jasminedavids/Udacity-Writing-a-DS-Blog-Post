```python
import pandas as pd
import pandasql as ps
import numpy as np
import matplotlib as plt 
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
%matplotlib inline
# Read meat consumption data file in
meat_consump = pd.read_csv('./meat_consumption.csv')
```

It is clear from the column variables that the data needs to be cleaned. First let's remove all of the constant and completely NaN columns as they don't add anything to the analysis.

```python
# Shows that the whole Flag Codes col is NaN values
print("Flag codes is all null: " + str(meat_consump['Flag Codes'].isnull().sum()==meat_consump.shape[0]))
print("Unnamed: 2 is all null: " + str(meat_consump['Unnamed: 2'].isnull().sum()==meat_consump.shape[0]))

# Shows that the INDICATOR and FREQUENCY cols are both constant
print("All of INDICATOR's values: " + str(np.unique(meat_consump['INDICATOR'])))
print("All of FREQUENCY's values: " + str(np.unique(meat_consump['FREQUENCY'])))
```

```python
# Removes Flag Codes, Unnamed:2, INDICATOR, and FREQUENCY cols
meat_consump.pop('Flag Codes')
meat_consump.pop("Unnamed: 2")
meat_consump.pop('INDICATOR')
meat_consump.pop('FREQUENCY')
meat_consump.head()
```

```python
# Verify that missing data is handled
print(meat_consump.isnull().sum())

# Make heading style clear and consistent
meat_consump = meat_consump.rename(columns = {'Tons of Beef': 'beef_tons'})
meat_consump = meat_consump.rename(columns = {'Tons of Pork': 'pork_tons'})
meat_consump = meat_consump.rename(columns = {'Tons of Poultry': 'poultry_tons'})
meat_consump = meat_consump.rename(columns = {'Tons of Sheep': 'sheep_tons'})
meat_consump = meat_consump.rename(columns = {'Beef kg/cap': 'beef_kg_cap'})
meat_consump = meat_consump.rename(columns = {'Pork kg/cap': 'pork_kg_cap'})
meat_consump = meat_consump.rename(columns = {'Poultry kg/cap': 'poultry_kg_cap'})
meat_consump = meat_consump.rename(columns = {'Sheep kg/cap': 'sheep_kg_cap'})
meat_consump = meat_consump.rename(columns = {'LOCATION': 'country'})
meat_consump = meat_consump.rename(columns = {'TIME': 'year'})
```

```python
# Create variable for cumulative consumption in tons and cumulative meat consumption in kg/capita
meat_consump["tons_meat_total"] = meat_consump["beef_tons"]+meat_consump["pork_tons"]+meat_consump["poultry_tons"]+meat_consump["sheep_tons"]
meat_consump["total_meat_kg_cap"] = meat_consump["beef_kg_cap"]+meat_consump["pork_kg_cap"]+meat_consump["poultry_kg_cap"]+meat_consump["sheep_kg_cap"]
```

### Which country has seen the greatest average percent increase in meat consumption per capita per year?

```python
meat_consump
```

```python
#percent annual increase
pct_annual_increase =[]       
for x in range(meat_consump.shape[0]):
        if x==0:
            pct_annual_increase.append(0)
        else:
            if meat_consump['country'][x] == meat_consump['country'][x-1]:
                pct_annual_increase.append(100*((meat_consump['total_meat_kg_cap'][x])/(meat_consump['total_meat_kg_cap'][x-1]) - 1))
            else: pct_annual_increase.append(0)
                
  
            
```

```python
meat_consump['percent_annual_increase'] = pd.Series(pct_annual_increase)
```
